```python
import numpy as np
import pandas as pd

# Stub until the package is published — shows the expected API surface.
# Swap for `from baselion_stress import StressSuite, load_ohlcv` once live.

rng = np.random.default_rng(42)
n = 24 * 365 * 2  # 2 years of hourly bars
price = 30000 * np.exp(np.cumsum(rng.normal(0, 0.01, n)))
prices = pd.Series(price, index=pd.date_range('2024-01-01', periods=n, freq='h'), name='close')
prices.head()
```

```python
def sma_crossover(prices, fast=20, slow=100):
    fast_ma = prices.rolling(fast).mean()
    slow_ma = prices.rolling(slow).mean()
    return np.sign(fast_ma - slow_ma).fillna(0.0)

signal = sma_crossover(prices)
returns = prices.pct_change().shift(-1) * signal
sharpe = returns.mean() / returns.std() * np.sqrt(24 * 365)
print(f'naive Sharpe (no stress test): {sharpe:.2f}')
```
